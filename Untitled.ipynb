```python
import itertools
from sklearn.metrics import confusion_matrix

# Function to plot Confusion Matrix
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion Matrix', cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.figure(figsize=(6, 6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=0)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()

```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Train Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Train Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Store models in a dictionary
models = {
    "Decision Tree": dt_model,
    "Random Forest": rf_model
}

# Function to plot confusion matrix
def plot_confusion_matrix(cm, classes, title="Confusion Matrix"):
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(title)
    plt.show()

# Confusion Matrix for Decision Tree
y_pred_dt = models["Decision Tree"].predict(X_test)
cm_dt = confusion_matrix(y_test, y_pred_dt)

print("Confusion Matrix - Decision Tree")
print(cm_dt)
plot_confusion_matrix(cm_dt, classes=[0, 1], title="Confusion Matrix - Decision Tree")

```

```python
from collections import Counter
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_confusion_matrix(cm, classes, title):
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(title)
    plt.show()
```

```python
def metrics(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    print(f'Accuracy: {accuracy:.2f}')
    print(f'Precision: {precision:.2f}')
    print(f'Recall: {recall:.2f}')
    print(f'F1 Score: {f1:.2f}')
```

```python

# Define features (X) and target variable (Y)
X = data.drop(columns=["Class"])  # Assuming "Class" is the target column
Y = data["Class"]  # The target variable

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, Y_resampled = smote.fit_resample(X, Y)

# Split the resampled data
X_train_resampled, X_test_resampled, y_train_resampled, y_test_resampled = train_test_split(
    X_resampled, Y_resampled, test_size=0.2, random_state=42, stratify=Y_resampled
)

print("Resampled shape of X:", X_resampled.shape)
print("Resampled shape of Y:", Y_resampled.shape)

value_counts = Counter(Y_resampled)
print(value_counts)

```

```python
rf_model_resampled = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model_resampled.fit(X_train_resampled, y_train_resampled)
```

```python
# Ensure you have training and test data
rf_resampled = RandomForestClassifier()
rf_resampled.fit(X_train_resampled, y_train_resampled)  # Train the model

# Predictions and model evaluation
predictions_resampled = rf_resampled.predict(X_test)
random_forest_score_resampled = rf_resampled.score(X_test, y_test) * 100

print(f"Random Forest Model Accuracy: {random_forest_score_resampled:.2f}%")
```

```python
# Confusion matrix visualization
y_pred_rf_resampled = rf_model_resampled.predict(X_test_resampled)
cm_rf_resampled = confusion_matrix(y_test_resampled, y_pred_rf_resampled)

print("Confusion Matrix - Random Forest After Oversampling")
print(cm_rf_resampled)
plot_confusion_matrix(cm_rf_resampled, classes=[0, 1], title="Confusion Matrix - Random Forest After Oversampling")
```

```python
# Evaluation
print("Evaluation of Random Forest Model")
metrics(y_test, predictions_resampled)
```

### 15. Graph Based Fraud Detection

```python
# Graph-Based Fraud Detection for Credit Card Transactions
def build_transaction_graph(data):
    G = nx.Graph()
    
    for index, row in data.iterrows():
        sender = f"User_{index}"  # Create synthetic user ID
        receiver = f"Merchant_{index}"  # Simulate merchant ID
        amount = row['Amount']
        fraud = row['Class']
        
        G.add_edge(sender, receiver, weight=amount)
        G.nodes[sender]['fraud'] = fraud
        G.nodes[receiver]['fraud'] = fraud
    
    return G

# Create transaction graph
graph = build_transaction_graph(data)

# Visualizing the network
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(graph, seed=42)
nx.draw(graph, pos, node_size=50, with_labels=False, edge_color='gray')
plt.title("Credit Card Transaction Network")
plt.show()

# Identify high-risk nodes (fraudulent clusters)
fraudulent_nodes = [n for n, attr in graph.nodes(data=True) if 'fraud' in attr and attr['fraud'] == 1]
nx.draw_networkx_nodes(graph, pos, nodelist=fraudulent_nodes, node_color='red', label='Fraudulent')
nx.draw_networkx_edges(graph, pos, alpha=0.5)
plt.title("Fraudulent Credit Card Transactions")
plt.legend()
plt.show()

# Additional Visualization: Transaction Amount Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data['Amount'], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()
```
